```python

############################### Comvertir y guardar .txt a csv ###############################

# Función para procesar una línea del archivo .txt
def procesar_linea(linea):
    valores = linea.strip().split(',')
    return [float(val) for val in valores]

# Función para leer el archivo .txt y devolver una lista de listas
def leer_archivo_txt(ruta):
    with open(ruta, 'r') as f:
        lineas = f.readlines()
        return [procesar_linea(linea) for linea in lineas]

# Directorio donde se encuentran los archivos .tar
directorio_base = 'Secuencias_neg'

# Obtener lista de archivos .tar en el directorio
archivos_tar = [f for f in os.listdir(directorio_base) if f.endswith('.tar')]

# Iterar sobre los archivos .tar
for archivo_tar in archivos_tar:
    # Ruta completa del archivo .tar
    ruta_tar = os.path.join(directorio_base, archivo_tar)
    
    # Extraer el archivo .txt del .tar
    with tarfile.open(ruta_tar, 'r') as tar:
        nombre_txt = 'training_code_normalized.txt'  # Cambia el nombre si es diferente
        miembro = tar.getmember(nombre_txt)
        tar.extract(miembro, path=directorio_base)
        ruta_txt = os.path.join(directorio_base, nombre_txt)

    # Leer el archivo .txt y procesar los datos
    datos = leer_archivo_txt(ruta_txt)
    df = pd.DataFrame(datos)

    # Construir la ruta para el nuevo archivo .csv
    nombre_csv = 'Feature_normalization_training.csv'  # Cambia el nombre si es diferente
    ruta_csv = os.path.join(directorio_base, nombre_csv)

    # Guardar el DataFrame como archivo CSV
    df.to_csv(ruta_csv, index=False, header=False)

    # Eliminar el archivo .txt ya que no es necesario mantenerlo
    os.remove(ruta_txt)

    # Agregar el archivo .csv al .tar
    with tarfile.open(ruta_tar, 'a') as tar:
        arcname = nombre_csv
        tar.add(ruta_csv, arcname=arcname)

    # Eliminar el archivo .csv ya que no es necesario mantenerlo
    os.remove(ruta_csv)

    print(f"Proceso completado para {archivo_tar}.")

print("Todos los archivos .csv han sido agregados a los respectivos .tar.")
```
