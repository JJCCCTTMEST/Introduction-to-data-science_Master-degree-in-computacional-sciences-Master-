```python
import tarfile
import os
import csv
import numpy as np
import pandas as pd
```

```python
########### Leer el contenido de cada csv en el archivo .tar##################


def extraer_csv_de_tar(archivo_tar, nombre_csv, directorio_destino):
    with tarfile.open(archivo_tar, 'r') as tar:
        miembro = tar.getmember(nombre_csv)
        tar.extract(miembro, path=directorio_destino)
        return os.path.join(directorio_destino, nombre_csv)

def contar_columnas_con_datos(ruta_csv):
    # Leer el archivo CSV
    df = pd.read_csv(ruta_csv)
    
    # Obtener las columnas con datos
    columnas_con_datos = df.columns[df.count() > 0].tolist()
    
    # Contar el total de columnas
    total_columnas = len(df.columns)
    
    return columnas_con_datos, total_columnas

# Nombre del archivo CSV a extraer (igual en todos los .tar)
nombre_csv = 'Feature_normalization_training.csv'

# Directorio donde se encuentran los archivos .tar
directorio_base = 'C:/Users/ADMIN/Secuencias_pos'

# Obtener lista de archivos .tar en el directorio
archivos_tar = [f for f in os.listdir(directorio_base) if f.endswith('.tar')]

for archivo_tar in archivos_tar:
    # Ruta completa del archivo .tar
    ruta_tar = os.path.join(directorio_base, archivo_tar)
    
    # Extraer el archivo CSV del .tar
    ruta_csv_extraido = extraer_csv_de_tar(ruta_tar, nombre_csv, directorio_base)

    # Llamada a la función para obtener las columnas con datos y el total de columnas
    columnas_con_datos, total_columnas = contar_columnas_con_datos(ruta_csv_extraido)

    print(f'Para el archivo .tar "{archivo_tar}":')
    print(f'  - El archivo CSV "{ruta_csv_extraido}" tiene un total de {total_columnas} columnas.')
    print(f'  - De estas, {len(columnas_con_datos)} tienen datos:')
```

```python

def extraer_csv_de_tar(archivo_tar, nombre_csv, directorio_destino):
    with tarfile.open(archivo_tar, 'r') as tar:
        miembro = tar.getmember(nombre_csv)
        tar.extract(miembro, path=directorio_destino)
        return os.path.join(directorio_destino, nombre_csv)

def contar_columnas_con_datos(ruta_csv):
    # Leer el archivo CSV
    df = pd.read_csv(ruta_csv)
    
    # Obtener las columnas con datos
    columnas_con_datos = df.columns[df.count() > 0].tolist()
    
    # Contar el total de columnas
    total_columnas = len(df.columns)
    
    return columnas_con_datos, total_columnas

# Nombre del archivo CSV a extraer (igual en todos los .tar)
nombre_csv = 'Feature_normalization_training.csv'

# Directorio donde se encuentran los archivos .tar
directorio_base = 'C:/Users/ADMIN/Secuencias_neg'

# Obtener lista de archivos .tar en el directorio
archivos_tar = [f for f in os.listdir(directorio_base) if f.endswith('.tar')]

# Lista para almacenar los resultados
resultados = []

for archivo_tar in archivos_tar:
    # Ruta completa del archivo .tar
    ruta_tar = os.path.join(directorio_base, archivo_tar)
    
    # Extraer el archivo CSV del .tar
    ruta_csv_extraido = extraer_csv_de_tar(ruta_tar, nombre_csv, directorio_base)

    # Llamada a la función para obtener las columnas con datos y el total de columnas
    columnas_con_datos, total_columnas = contar_columnas_con_datos(ruta_csv_extraido)

    # Almacenar resultados en la lista
    resultados.append({
        'Archivo_tar': archivo_tar,
        'Ruta_CSV': ruta_csv_extraido,
        'Total_Columnas': total_columnas,
        'Columnas_Con_Datos': len(columnas_con_datos)
    })

# Convertir la lista de resultados a un DataFrame
resultados_df = pd.DataFrame(resultados)

# Almacenar el DataFrame en un archivo CSV
resultados_df.to_csv('Columnas en .tar negativos.csv', index=False)

print('Resultados almacenados en "resultados.csv".')
```
